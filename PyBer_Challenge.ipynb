```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Files to load
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"
```

```python
# Read the city data file and store it in a pandas DataFrame.
city_data_df = pd.read_csv(city_data_to_load)
city_data_df.head(10)
```

```python
# Read the ride data file and store it in a pandas DataFrame.
ride_data_df = pd.read_csv(ride_data_to_load)
ride_data_df.head(10)
```

```python
# Get the columns and the rows that are not null.
city_data_df.count()
```

```python
# Get the columns and the rows that are not null.
ride_data_df.count()
```

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the DataFrame
pyber_data_df.head()
```

```python
# Total rides by type
total_rides_by_type = pyber_data_df.groupby("type").count()["ride_id"]
total_rides_by_type
```

```python
#Total drivers by city type
total_drivers_by_type = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_by_type
```

```python
#Total fares by type
total_fares_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_by_type                                                     
```

```python
#Average fare per ride for cities
avg_fare_per_ride_by_type = total_fares_by_type / total_rides_by_type 
avg_fare_per_ride_by_type
```

```python
#Average fare per driver for cities 
avg_fare_per_driver_by_type = total_fares_by_type / total_drivers_by_type 
avg_fare_per_driver_by_type
```

```python
# Create summary 
pyber_summary_df = pd.DataFrame({
    "Total Rides": total_rides_by_type,
    "Total Drivers": total_drivers_by_type,
    "Total Fares": total_fares_by_type,
    "Average Fare per Ride": avg_fare_per_ride_by_type,
    "Average Fare per Driver": avg_fare_per_driver_by_type})
pyber_summary_df
```

```python
#Delete index name 
pyber_summary_df.index.name = None
```

```python
pyber_summary_df
```

```python
#Formatting 
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)

pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)

pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:.2f}".format)

pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)

pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)
pyber_summary_df
```

```python
#Rename the merged dataframe columns 
pyber_data_df = pyber_data_df.rename(columns={'city':'City', 'date':'Date', 'fare':'Fare', 'ride_id':'Ride Id', 'driver_count':'No. Drivers', 'type':'City Type'})
pyber_data_df.head()
```

```python
# set index to date column
pyber_data_df.set_index(pyber_data_df['Date'], inplace=True)
pyber_data_df
```

```python
#Make copy of merged dataframe 
pyber_data_copy = pyber_data_df[['Date', 'City Type', 'Fare']].copy()
pyber_data_copy
```

```python
#Drop the extra date column
pyber_data_copy.drop(['Date'], axis=1, inplace=True)
```

```python
#Change to datetime format
pyber_data_copy.index = pd.to_datetime(pyber_data_df.index)
pyber_data_copy.head(10)
```

```python
#Use info to check index 
pyber_data_copy.info()
```

```python
#Calculate total fares by city and date 
total_fares_by_type_date_df = pyber_data_copy.groupby(["City Type", "Date"]).sum()["Fare"]
total_fares_by_type_date_df.head(10)
```

```python
#Convert the groupby into a dataframe
total_fares_by_type_date_df = pd.DataFrame(total_fares_by_type_date_df)
total_fares_by_type_date_df.head(10)
```

```python
#Reset index
total_fares_by_type_date_df = total_fares_by_type_date_df.reset_index()
total_fares_by_type_date_df.head(10)
total_fares_by_type_date_df.info()
```

```python
total_fares_by_type_date_df.rename_axis(None, inplace=True)
```

```python
#Make pivot table 
total_fares_pivot = total_fares_by_type_date_df.pivot(index='Date', columns='City Type')['Fare']
total_fares_pivot.head(10)
```

```python
total_fares_pivot.info()
```

```python
#Use loc to group by 01-04 
jan_april_df = total_fares_pivot.loc['2019-01-01':'2019-04-28']
jan_april_df.info()
jan_april_df.head()
```

```python
#Create weekly bins resampled dataframe 
weekly_resampled_jan_april_df = jan_april_df.resample('W').sum()
weekly_resampled_jan_april_df
```

```python
import matplotlib.style as style
```
